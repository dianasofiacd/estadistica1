```python
# =====================================================
# Intervalo de confianza para la media poblacional
# =====================================================

# Librerías necesarias
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# -----------------------------------------------------
# 1. Planteamiento del problema
# -----------------------------------------------------
# Queremos estimar la altura promedio de estudiantes universitarios.
# Datos conocidos:
# n = 50 (tamaño de la muestra)
# x̄ = 170 (media muestral)
# σ = 10 (desviación estándar poblacional conocida)
# Nivel de confianza = 95% (α = 0.05)

n = 50
xbar = 170
sigma = 10
alpha = 0.05

# -----------------------------------------------------
# 2. Fórmula del intervalo de confianza
# -----------------------------------------------------
# IC = x̄ ± Z_(α/2) * (σ / √n)
# Usamos distribución normal porque σ es conocida.

# Valor crítico Z para 95% de confianza
z_crit = stats.norm.ppf(1 - alpha/2)

# Error estándar de la media
se = sigma / np.sqrt(n)

# Margen de error
me = z_crit * se

# Límites del intervalo de confianza
li = xbar - me
ls = xbar + me

print("Intervalo de confianza del 95% para la media poblacional:")
print(f"({li:.2f}, {ls:.2f})")

# -----------------------------------------------------
# 3. Interpretación
# -----------------------------------------------------
print("\nInterpretación:")
print(f"Con un 95% de confianza, se estima que la media poblacional de la altura")
print(f"se encuentra entre {li:.2f} cm y {ls:.2f} cm.")
print("Esto significa que si repitiésemos este muestreo muchas veces, en el 95% de los casos")
print("el intervalo incluiría la verdadera media poblacional.")

# -----------------------------------------------------
# 4. Gráfico de la distribución normal
# -----------------------------------------------------
# Graficamos la distribución normal centrada en la media muestral con el IC sombreado.

x = np.linspace(xbar - 4*se, xbar + 4*se, 1000)
y = stats.norm.pdf(x, xbar, se)

plt.figure(figsize=(10,6))
plt.plot(x, y, label="Distribución de la media muestral")

# Sombrear el intervalo de confianza
plt.fill_between(x, y, where=(x >= li) & (x <= ls), color="skyblue", alpha=0.5, label="IC 95%")

# Líneas de los límites
plt.axvline(li, color="red", linestyle="--", label=f"Límite inferior = {li:.2f}")
plt.axvline(ls, color="red", linestyle="--", label=f"Límite superior = {ls:.2f}")

# Línea de la media muestral
plt.axvline(xbar, color="black", linestyle="-.", label=f"Media muestral = {xbar:.2f}")

# Añadir título y etiquetas
plt.title("Distribución de la Media Muestral e Intervalo de Confianza del 95%")
plt.xlabel("Altura (cm)")
plt.ylabel("Densidad de Probabilidad")
plt.legend()
plt.grid(True)
plt.show()
```

11. El departamento de salud planea una encuesta para estimar la proporción de residentes vacunados
contra la influenza. En un estudio piloto, el 72 % de una muestra mostró estar vacunado. Determine
el tamaño de muestra necesario para estimar esta proporción con un margen de error del 3 % y un
nivel de confianza del 95 %. Si no se tuviera el estudio piloto, ¿cuál sería el tamaño requerido?

```python
import math
from scipy import stats

# Datos del problema
p_hat = 0.72      # proporción piloto
E = 0.03          # margen de error (3%)
conf = 0.95       # nivel de confianza
alpha = 1 - conf

# Valor crítico Z para dos colas
z = stats.norm.ppf(1 - alpha/2)

# Tamaño de muestra usando p_hat del estudio piloto
n_raw_pilot = (z**2 * p_hat * (1 - p_hat)) / (E**2)
n_pilot = math.ceil(n_raw_pilot)

# Tamaño de muestra si no se tuviera estudio piloto (p = 0.5 -> máxima varianza)
p_worst = 0.5
n_raw_worst = (z**2 * p_worst * (1 - p_worst)) / (E**2)
n_worst = math.ceil(n_raw_worst)

print(f"Nivel de confianza: {conf*100:.0f}%  |  Z = {z:.4f}  |  Margen de error = {E:.3f}")
print("\nUsando p̂ = 0.72 (estudio piloto):")
print(f"  n (exacto) = {n_raw_pilot:.3f}")
print(f"  n (redondeado hacia arriba) = {n_pilot}")

print("\nSi no hay estudio piloto (usar p = 0.5):")
print(f"  n (exacto) = {n_raw_worst:.3f}")
print(f"  n (redondeado hacia arriba) = {n_worst}")

# Opcional: si se conoce población finita N, descomentar y setear N para aplicar corrección:
# N = 10000  # ejemplo
# n_finite = math.ceil((n_pilot * N) / (n_pil_*
```

21. Una empresa afirma que sus paquetes de cereal tienen un peso promedio de 500 g. Se sospecha que
la máquina está descalibrada y subllenando. Se toma una muestra de 40 paquetes, obteniendo una
media de 495 g. La desviación estándar poblacional es σ = 15 g. Con un nivel de de significancia
α = 0,05.
1. Plantee las hipótesis nula y alternativa.
2. Calcule el estadístico Z.
3. Determine la región crítica y la región de rechazo.
4. Concluya si se rechaza H0.


```python
# Ejercicio 21: Prueba de hipótesis (cola izquierda) con sigma conocida
# 1) Plantear H0 y Ha
# H0: mu = 500 (la máquina está correcta)
# Ha: mu < 500 (la máquina está subllenando) --> prueba unilateral izquierda

import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import math

# Datos del problema
mu0 = 500.0      # valor hipotético bajo H0 (g)
xbar = 495.0     # media muestral (g)
sigma = 15.0     # desviación estándar poblacional (g)
n = 40           # tamaño de la muestra
alpha = 0.05     # nivel de significancia

# 2) Estadístico Z
se = sigma / math.sqrt(n)                       # error estándar
z_stat = (xbar - mu0) / se                      # estadístico Z
# Para prueba unilateral izquierda, el p-valor es la cola izquierda de Z
p_value = stats.norm.cdf(z_stat)

# 3) Región crítica (cola izquierda)
z_crit = stats.norm.ppf(alpha)                  # valor crítico para cola izquierda
reject = z_stat < z_crit                        # condición para rechazar H0

# Mostrar resultados numéricos
print(f"Planteamiento:")
print(" H0: μ = {:.1f} g".format(mu0))
print(" Ha: μ < {:.1f} g".format(mu0))
print()
print(f"Parámetros muestrales: n = {n}, x̄ = {xbar} g, σ = {sigma} g")
print(f"Nivel de significancia: α = {alpha}")
print()
print("Cálculos:")
print(f" Error estándar (σ/√n): {se:.4f}")
print(f" Estadístico Z: {z_stat:.4f}")
print(f" Valor crítico (Z_crit para α={alpha}): {z_crit:.4f}")
print(f" p-valor (cola izquierda): {p_value:.6f}")
print()
print("Decisión:")
if reject:
    print(" Rechazamos H0 (el estadístico está en la región de rechazo).")
else:
    print(" No se rechaza H0 (el estadístico NO está en la región de rechazo).")

# Interpretación práctica
if reject:
    print("\nConclusión práctica: Hay evidencia estadística, al nivel α = {:.2f}, de que la máquina está subllenando (el peso promedio es menor que 500 g).".format(alpha))
else:
    print("\nConclusión práctica: No hay evidencia suficiente, al nivel α = {:.2f}, para afirmar que la máquina está subllenando; no se rechaza que el peso promedio sea 500 g.".format(alpha))
print("\nSignificado del p-valor: Probabilidad de observar un estadístico tan extremo (o más) en la cola izquierda, asumiendo que H0 es verdadera.")

# 6) Gráfica: distribución normal estándar con región de rechazo y z observado
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(9,5))
plt.plot(x, y, label='Distribución Normal estándar N(0,1)')

# Región de rechazo (cola izquierda)
plt.fill_between(x, 0, y, where=(x < z_crit), color='red', alpha=0.3, label='Región de rechazo (cola izquierda)')

# Líneas verticales: z_crit y z_stat
plt.axvline(z_crit, color='red', linestyle='--', label=f'Z crítico = {z_crit:.3f}')
plt.axvline(z_stat, color='blue', linestyle='-', label=f'Z observado = {z_stat:.3f}')

# Marcar p-valor area (área a la izquierda de z_stat)
x_fill = np.linspace(-4, z_stat, 200)
plt.fill_between(x_fill, 0, stats.norm.pdf(x_fill), color='blue', alpha=0.2, label=f'Área p-valor = {p_value:.4f}')

plt.title('Prueba Z (cola izquierda) — H0: μ = 500 g vs Ha: μ < 500 g')
plt.xlabel('Z')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ejercicio 22: Prueba de hipótesis (cola izquierda) con sigma conocida
# Caso: fabricante de baterías

import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import math

# --- Datos del problema ---
mu0 = 1200.0    # valor hipotético bajo H0 (horas)
xbar = 1180.0   # media muestral (horas)
sigma = 100.0   # desviación estándar poblacional (horas)
n = 50          # tamaño de la muestra
alpha = 0.01    # nivel de significancia

# --- 1) Planteamiento de hipótesis ---
# H0: μ = 1200 (la duración promedio es correcta)
# Ha: μ < 1200 (la duración es menor) --> prueba unilateral izquierda

# --- 2) Estadístico Z ---
se = sigma / math.sqrt(n)         # error estándar
z_stat = (xbar - mu0) / se        # estadístico Z
p_value = stats.norm.cdf(z_stat)  # p-valor cola izquierda

# --- 3) Región crítica ---
z_crit = stats.norm.ppf(alpha)    # valor crítico para α = 0.01
reject = z_stat < z_crit

# --- Resultados ---
print("Planteamiento:")
print(f" H0: μ = {mu0} horas")
print(f" Ha: μ < {mu0} horas")
print()
print(f"Datos: n = {n}, x̄ = {xbar} horas, σ = {sigma} horas")
print(f"Nivel de significancia α = {alpha}")
print()
print("Cálculos:")
print(f" Error estándar (σ/√n): {se:.4f}")
print(f" Estadístico Z: {z_stat:.4f}")
print(f" Valor crítico Z (α={alpha}): {z_crit:.4f}")
print(f" p-valor: {p_value:.6f}")
print()
print("Decisión:")
if reject:
    print(" Rechazamos H0: el estadístico está en la región de rechazo.")
    print(" Conclusión práctica: Hay evidencia al nivel α = 0.01 de que la duración promedio es menor a 1200 horas.")
else:
    print(" No se rechaza H0: el estadístico NO cae en la región de rechazo.")
    print(" Conclusión práctica: No hay suficiente evidencia al nivel α = 0.01 para afirmar que la duración es menor a 1200 horas.")

# --- 4) Gráfico de la distribución normal estándar ---
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(9,5))
plt.plot(x, y, label='Distribución Normal estándar N(0,1)')

# Región de rechazo en cola izquierda
plt.fill_between(x, 0, y, where=(x < z_crit), color='red', alpha=0.3, label='Región de rechazo')

# Líneas de referencia
plt.axvline(z_crit, color='red', linestyle='--', label=f'Z crítico = {z_crit:.3f}')
plt.axvline(z_stat, color='blue', linestyle='-', label=f'Z observado = {z_stat:.3f}')

# Área del p-valor
x_fill = np.linspace(-4, z_stat, 200)
plt.fill_between(x_fill, 0, stats.norm.pdf(x_fill), color='blue', alpha=0.2, label=f'Área p-valor = {p_value:.4f}')

plt.title('Prueba Z unilateral (cola izquierda)\nH0: μ = 1200 vs Ha: μ < 1200')
plt.xlabel('Z')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ejercicio 23: Prueba de hipótesis bilateral con sigma conocida
# Caso: promedio de clase vs promedio nacional

import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import math

# --- Datos del problema ---
mu0 = 75.0      # valor hipotético bajo H0 (promedio nacional)
xbar = 78.0     # media muestral (promedio clase)
sigma = 12.0    # desviación estándar poblacional
n = 36          # tamaño de la muestra
alpha = 0.05    # nivel de significancia

# --- 1) Hipótesis ---
# H0: μ = 75
# Ha: μ ≠ 75  (bilateral)

# --- 2) Estadístico Z ---
se = sigma / math.sqrt(n)         # error estándar
z_stat = (xbar - mu0) / se        # estadístico Z
# p-valor bilateral = 2 * cola superior de |z|
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# --- 3) Región crítica ---
z_crit = stats.norm.ppf(1 - alpha/2)   # valor crítico para cola bilateral
reject = abs(z_stat) > z_crit

# --- Resultados ---
print("Planteamiento:")
print(f" H0: μ = {mu0}")
print(f" Ha: μ ≠ {mu0}")
print()
print(f"Datos: n = {n}, x̄ = {xbar}, σ = {sigma}")
print(f"Nivel de significancia α = {alpha}")
print()
print("Cálculos:")
print(f" Error estándar (σ/√n): {se:.4f}")
print(f" Estadístico Z: {z_stat:.4f}")
print(f" Valor crítico bilateral ±Z_crit: ±{z_crit:.4f}")
print(f" p-valor: {p_value:.6f}")
print()
print("Decisión:")
if reject:
    print(" Rechazamos H0: el estadístico está en la región de rechazo.")
    print(" Conclusión práctica: Hay evidencia al nivel α = 0.05 de que el promedio de la clase es diferente a 75.")
else:
    print(" No se rechaza H0: el estadístico NO cae en la región de rechazo.")
    print(" Conclusión práctica: No hay suficiente evidencia al nivel α = 0.05 para afirmar que el promedio difiera de 75.")

# --- 4) Gráfico ---
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(9,5))
plt.plot(x, y, label='Distribución Normal estándar N(0,1)')

# Regiones de rechazo (bilateral)
plt.fill_between(x, 0, y, where=(x < -z_crit) | (x > z_crit), color='red', alpha=0.3, label='Regiones de rechazo')

# Líneas de referencia
plt.axvline(-z_crit, color='red', linestyle='--', label=f'-Z crítico = {-z_crit:.3f}')
plt.axvline(z_crit, color='red', linestyle='--', label=f'Z crítico = {z_crit:.3f}')
plt.axvline(z_stat, color='blue', linestyle='-', label=f'Z observado = {z_stat:.3f}')

# Área p-valor (bilateral)
x_fill_left = np.linspace(-4, -abs(z_stat), 200)
plt.fill_between(x_fill_left, 0, stats.norm.pdf(x_fill_left), color='blue', alpha=0.2)
x_fill_right = np.linspace(abs(z_stat), 4, 200)
plt.fill_between(x_fill_right, 0, stats.norm.pdf(x_fill_right), color='blue', alpha=0.2, label=f'p-valor = {p_value:.4f}')

plt.title('Prueba Z bilateral\nH0: μ = 75 vs Ha: μ ≠ 75')
plt.xlabel('Z')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ejercicio 26: Prueba t unilateral (cola derecha) - tiempo de entrega
# H0: μ = 25 (minutos)
# Ha: μ > 25 (el tiempo de entrega es mayor) --> prueba unilateral derecha

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import math

# --- Datos ---
mu0 = 25.0        # valor bajo H0 (minutos)
xbar = 28.0       # media muestral (minutos)
s = 4.0           # desviación estándar muestral
n = 15            # tamaño de muestra
alpha = 0.05      # nivel de significancia

# --- 2) Estadístico t ---
se = s / math.sqrt(n)               # error estándar de la media (usando s)
t_stat = (xbar - mu0) / se          # estadístico t

# --- 3) Grados de libertad y valor crítico (cola derecha) ---
df = n - 1
t_crit = stats.t.ppf(1 - alpha, df)   # valor crítico para prueba unilateral derecha

# p-valor para prueba unilateral derecha
p_value = 1 - stats.t.cdf(t_stat, df)

# --- Resultados numéricos ---
print("Hipótesis:")
print(f" H0: μ = {mu0}")
print(f" Ha: μ > {mu0} (cola derecha)")
print()
print(f"Datos: n = {n}, x̄ = {xbar}, s = {s}")
print(f"Nivel de significancia: α = {alpha}")
print()
print("Cálculos:")
print(f" Error estándar (s/√n): {se:.4f}")
print(f" Estadístico t: {t_stat:.4f}")
print(f" Grados de libertad: {df}")
print(f" Valor crítico t (unilateral, α={alpha}): {t_crit:.4f}")
print(f" p-valor (cola derecha): {p_value:.6f}")
print()

# Decisión
if t_stat > t_crit:
    print("Decisión: Rechazamos H0 (t está en la región de rechazo).")
    print("Conclusión práctica: Hay evidencia, al nivel α = {:.2f}, de que el tiempo promedio de entrega es mayor a 25 minutos.".format(alpha))
else:
    print("Decisión: No se rechaza H0 (t NO está en la región de rechazo).")
    print("Conclusión práctica: No hay evidencia suficiente, al nivel α = {:.2f}, para afirmar que el tiempo promedio sea mayor a 25 minutos.".format(alpha))

# --- 4) Gráfica de la distribución t con región de rechazo y t observado ---
x_min = -4
x_max = max(6, t_stat + 1)
x = np.linspace(x_min, x_max, 1000)
y = stats.t.pdf(x, df)

plt.figure(figsize=(9,5))
plt.plot(x, y, label=f"t-distribution (df={df})")

# Región de rechazo (cola derecha)
plt.fill_between(x, 0, y, where=(x >= t_crit), color='red', alpha=0.3, label='Región de rechazo (cola derecha)')

# Área p-valor (from t_stat to +inf)
x_fill = np.linspace(t_stat, x_max, 300)
y_fill = stats.t.pdf(x_fill, df)
plt.fill_between(x_fill, 0, y_fill, color='blue', alpha=0.2, label=f'Área p-valor = {p_value:.4f}')

# Líneas verticales: t_crit y t_stat
plt.axvline(t_crit, color='red', linestyle='--', label=f't crítico = {t_crit:.3f}')
plt.axvline(t_stat, color='blue', linestyle='-', label=f't observado = {t_stat:.3f}')

plt.title('Prueba t (cola derecha)\nH0: μ = 25 vs Ha: μ > 25')
plt.xlabel('t')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ejercicio 29: Prueba t unilateral (cola izquierda)
# El suplemento promete un aumento promedio de μ0 = 2 mg/dL.
# Datos muestrales: n = 8, x̄ = 1.5 mg/dL, s = 0.6 mg/dL
# Nivel de significancia: α = 0.10
#
# Asumimos prueba unilateral izquierda porque la muestra muestra un aumento menor
# que el prometido y queremos comprobar si el verdadero aumento es menor a 2 mg/dL:
# H0: μ = 2
# Ha: μ < 2

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import math

# -----------------------
# Datos
mu0 = 2.0       # valor bajo H0 (mg/dL)
xbar = 1.5      # media muestral (mg/dL)
s = 0.6         # desviación estándar muestral (mg/dL)
n = 8           # tamaño de muestra
alpha = 0.10    # nivel de significancia
# -----------------------

# 2) Estadístico t
se = s / math.sqrt(n)            # error estándar (s/√n)
t_stat = (xbar - mu0) / se       # estadístico t

# 3) Grados de libertad y valor crítico (cola izquierda)
df = n - 1
t_crit = stats.t.ppf(alpha, df)  # valor crítico para cola izquierda

# p-valor (cola izquierda)
p_value = stats.t.cdf(t_stat, df)

# Imprimir resultados
print("Hipótesis:")
print(f" H0: μ = {mu0}")
print(f" Ha: μ < {mu0} (prueba unilateral, cola izquierda)")
print()
print(f"Datos: n = {n}, x̄ = {xbar}, s = {s}")
print(f"Nivel de significancia: α = {alpha}")
print()
print("Cálculos:")
print(f" Error estándar (s/√n): {se:.6f}")
print(f" Estadístico t: {t_stat:.6f}")
print(f" Grados de libertad: {df}")
print(f" Valor crítico t (cola izquierda, α={alpha}): {t_crit:.6f}")
print(f" p-valor (cola izquierda): {p_value:.6f}")
print()

# 4) Decisión
if t_stat < t_crit:
    print("✅ Decisión: Rechazamos H0 (t está en la región de rechazo).")
    print(f"Conclusión práctica: Al nivel α = {alpha}, hay evidencia de que el aumento promedio es menor a {mu0} mg/dL.")
else:
    print("❌ Decisión: No se rechaza H0.")
    print(f"Conclusión práctica: No hay evidencia suficiente, al nivel α = {alpha}, para afirmar que el aumento sea menor a {mu0} mg/dL.")

# -----------------------
# Gráfica: distribución t con región de rechazo y t observado
x_min = -4
x_max = max(4, t_stat + 2)
x = np.linspace(x_min, x_max, 1000)
y = stats.t.pdf(x, df)

plt.figure(figsize=(9,5))
plt.plot(x, y, label=f"t-distribución (df={df})")

# Región de rechazo (cola izquierda)
plt.fill_between(x, 0, y, where=(x <= t_crit), color='red', alpha=0.3, label='Región de rechazo (cola izquierda)')

# Área p-valor (desde -inf hasta t_stat)
x_fill = np.linspace(x_min, t_stat, 300)
y_fill = stats.t.pdf(x_fill, df)
plt.fill_between(x_fill, 0, y_fill, color='blue', alpha=0.2, label=f'Área p-valor = {p_value:.4f}')

# Líneas verticales: t_crit y t_stat
plt.axvline(t_crit, color='red', linestyle='--', label=f't crítico = {t_crit:.3f}')
plt.axvline(t_stat, color='blue', linestyle='-', label=f't observado = {t_stat:.3f}')

plt.title('Prueba t (cola izquierda)\nH0: μ = 2 mg/dL vs Ha: μ < 2 mg/dL')
plt.xlabel('t')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```

31. Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de
σ
2 = 2,5 minutos2
. Se toma una muestra de n = 16 llamadas y se obtiene una varianza muestral
de s
2 = 4. ¿Hay evidencia al 5 % de significancia de que la varianza poblacional difiere de 2.5?
Resuelva siguiendo el siguiente protocolo:
1. Plantee las hipótesis nula y alternativa.
2. Calcule el estadístico χ
2
.
3. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de
rechazo.
4. Concluya si se rechaza H0.


```python
# Ejercicio 31: Prueba de hipótesis para la varianza (χ²) — código listo para Google Colab
# H0: σ^2 = 2.5
# Ha: σ^2 ≠ 2.5  (prueba bilateral)
# Datos: n = 16, s^2 = 4, α = 0.05

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import math

# --- Datos ---
sigma0_sq = 2.5   # varianza bajo H0
s2 = 4.0          # varianza muestral
n = 16
alpha = 0.05

# --- 1) Hipótesis ---
# H0: sigma^2 = sigma0_sq
# Ha: sigma^2 != sigma0_sq (bilateral)

# --- 2) Estadístico chi-cuadrado ---
df = n - 1
chi2_stat = (df * s2) / sigma0_sq   # χ^2 = (n-1)*s^2 / σ0^2

# --- 3) Valores críticos para prueba bilateral ---
chi2_lower = stats.chi2.ppf(alpha/2, df)        # límite inferior
chi2_upper = stats.chi2.ppf(1 - alpha/2, df)    # límite superior

# p-valor (bilateral)
# Para χ^2 (right-skewed), p-value bilateral = 2 * min( P(χ^2 <= stat), P(χ^2 >= stat) )
cdf_stat = stats.chi2.cdf(chi2_stat, df)
p_value = 2 * min(cdf_stat, 1 - cdf_stat)

# --- Imprimir resultados ---
print("Hipótesis:")
print(f" H0: σ^2 = {sigma0_sq}")
print(" Ha: σ^2 ≠ {:.3f} (prueba bilateral)".format(sigma0_sq))
print()
print(f"Datos: n = {n}, s^2 = {s2}, df = {df}, α = {alpha}")
print()
print("Cálculos:")
print(f" Estadístico χ² = (n-1)*s² / σ0² = {chi2_stat:.4f}")
print(f" Valores críticos χ² (bilateral): lower = {chi2_lower:.4f}, upper = {chi2_upper:.4f}")
print(f" p-valor (bilateral) = {p_value:.6f}")
print()

# --- Decisión ---
if (chi2_stat < chi2_lower) or (chi2_stat > chi2_upper):
    print("Decisión: Rechazamos H0 (χ² está en la región de rechazo).")
    print("Conclusión práctica: Hay evidencia al nivel α = {:.2f} de que la varianza poblacional difiere de {:.3f}.".format(alpha, sigma0_sq))
else:
    print("Decisión: No se rechaza H0 (χ² NO está en la región de rechazo).")
    print("Conclusión práctica: No hay evidencia suficiente al nivel α = {:.2f} para afirmar que la varianza difiera de {:.3f}.".format(alpha, sigma0_sq))

# --- Gráfica: distribución χ² con regiones de rechazo y estadístico observado ---
x_min = 0
x_max = stats.chi2.ppf(0.999, df)
x = np.linspace(x_min, x_max, 1000)
y = stats.chi2.pdf(x, df)

plt.figure(figsize=(10,5))
plt.plot(x, y, label=f"χ² pdf (df={df})")

# Región de rechazo: left tail and right tail
plt.fill_between(x, 0, y, where=(x <= chi2_lower), color='red', alpha=0.3, label='Región de rechazo (cola izquierda)')
plt.fill_between(x, 0, y, where=(x >= chi2_upper), color='red', alpha=0.3, label='Región de rechazo (cola derecha)')

# Marca el estadístico observado
plt.axvline(chi2_stat, color='blue', linestyle='--', label=f'χ² observado = {chi2_stat:.4f}')
plt.axvline(chi2_lower, color='red', linestyle='--', label=f'χ² crítico inferior = {chi2_lower:.4f}')
plt.axvline(chi2_upper, color='red', linestyle='--', label=f'χ² crítico superior = {chi2_upper:.4f}')

# Sombrear área p-valor (both tails)
if chi2_stat <= (chi2_lower + chi2_upper)/2:
    # observed in left region or center -> shade left tail up to chi2_stat and mirror right tail
    x_fill_left = np.linspace(0, chi2_stat, 200)
    plt.fill_between(x_fill_left, 0, stats.chi2.pdf(x_fill_left, df), color='blue', alpha=0.2)
    # mirror area on right tail corresponding to same tail probability
    tail_prob = stats.chi2.cdf(chi2_stat, df)
    mirror_cut = stats.chi2.ppf(1 - tail_prob, df)
    x_fill_right = np.linspace(mirror_cut, x_max, 200)
    plt.fill_between(x_fill_right, 0, stats.chi2.pdf(x_fill_right, df), color='blue', alpha=0.2)
else:
    x_fill_right = np.linspace(chi2_stat, x_max, 200)
    plt.fill_between(x_fill_right, 0, stats.chi2.pdf(x_fill_right, df), color='blue', alpha=0.2)
    tail_prob = 1 - stats.chi2.cdf(chi2_stat, df)
    mirror_cut = stats.chi2.ppf(tail_prob, df)
    x_fill_left = np.linspace(0, mirror_cut, 200)
    plt.fill_between(x_fill_left, 0, stats.chi2.pdf(x_fill_left, df), color='blue', alpha=0.2)

plt.title('Prueba de varianza: distribución χ²\nH0: σ² = {:.3f} — prueba bilateral (α={})'.format(sigma0_sq, alpha))
plt.xlabel('χ²')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```
