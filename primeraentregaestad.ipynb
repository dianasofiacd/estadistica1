```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar archivo CSV
df = pd.read_csv("nacimientosinstisalud.csv", encoding='latin-1')

# Información básica
print("Información del dataset")
print(df.info())
print("Primeras filas del dataset")
print(df.head())
print("Resumen estadístico del dataset")
print(df.describe())
```

```python
# Valores faltantes
print("\nValores faltantes por columna:")
print(df.isnull().sum())

# Visualización de nulos
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes")
plt.show()
```

```python
df_filtrado_columnas=["PESO (Gramos)", "EDAD MADRE", "NÚMERO HIJOS NACIDOS VIVOS", "SEXO", "TIPO PARTO","ESTADO CONYUGAL MADRE"]
df_filtrado = df[df_filtrado_columnas]

print(df_filtrado.columns)
print("\nInformación del subset del dataset")
print(df_filtrado.info())
print("\nPrimeras filas del subset del dataset")
print(df_filtrado.head())
```

```python
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

numerical_cols = ["PESO (Gramos)", "NÚMERO HIJOS NACIDOS VIVOS", "EDAD MADRE"]
continuous_numerical_cols = ["PESO (Gramos)", "EDAD MADRE"]

for col in numerical_cols:
    # Q-Q plot
    plt.figure(figsize=(8, 6))
    sm.qqplot(df_filtrado[col], line='s')
    plt.title(f'Q-Q Plot de {col}')
    plt.show()

for col in continuous_numerical_cols:
    print(f"\nResultados de los tests de normalidad para '{col}':")

    # Shapiro-Wilk Test
    shapiro_test = stats.shapiro(df_filtrado[col])
    print(f"Shapiro-Wilk Test: Statistics={shapiro_test.statistic:.3f}, p-value={shapiro_test.pvalue:.3f}")

    # Kolmogorov-Smirnov Test (Lilliefors)
    kolmogorov_test = stats.kstest(df_filtrado[col], 'norm', args=(df_filtrado[col].mean(), df_filtrado[col].std()))
    print(f"Kolmogorov-Smirnov Test: Statistics={kolmogorov_test.statistic:.3f}, p-value={kolmogorov_test.pvalue:.3f}")

    # Anderson-Darling Test
    anderson_test = stats.anderson(df_filtrado[col], dist='norm')
    print(f"Anderson-Darling Test: Statistics={anderson_test.statistic:.3f}")
    for i in range(len(anderson_test.critical_values)):
        sl, cv = anderson_test.significance_level[i], anderson_test.critical_values[i]
        print(f'   Significance Level {sl:.1f}%: {cv:.3f} (Data is normal if A-D statistic < critical value)')

    # Jarque-Bera Test
    jarque_bera_test = stats.jarque_bera(df_filtrado[col])
    print(f"Jarque-Bera Test: Statistics={jarque_bera_test.statistic:.3f}, p-value={jarque_bera_test.pvalue:.3f}")
```
